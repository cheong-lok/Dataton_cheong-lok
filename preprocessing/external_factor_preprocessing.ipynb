```python
schoolViolence_1 = df_1.iloc[:, 6:16] # 학교폭력 피해경험
schoolViolence_2 = df_1.iloc[:, 24:26] # 학교폭력 기관 도움 여부, 정도 / 위와 합치기
sexualViolence_1 = df_1.iloc[:, 16:24] # 성폭력 피해경험
sexualViolence_2 = df_1.iloc[:, 26:28] # 성폭력 기관 도움 여부, 정도 / 위와 합치기

schoolViolence = pd.concat([schoolViolence_1, schoolViolence_2], axis=1) # 학교폭력 피해
sexualViolence = pd.concat([sexualViolence_1, sexualViolence_2], axis=1) # 성폭력 피해
entryRestricted = df_1.iloc[:, 29:] # 청소년 출입제한업소
loan = df_11.copy()
community = df_2.copy()
school_violence_help_1 = df_1.iloc[:, 0:6:2] # 기관 인지
school_violence_help_2 = df_1.iloc[:, 1:6:2] # 기관 이용 경험
school_violence_education = df_9[['T_106', 'T_107']]
sex_education = df_9[['T_112', 'T_113']]
media_education = df_9[['T_116', 'T_117', 'T_118', 'T_119', 'T_120', 'T_121', 'T_122', 'T_123']]
safety_education = df_9[['T_126']]
personal_edu = df_8.copy()
```

```python
a = reduce(lambda x,y: pd.merge(x,y, on='시도', how='outer'), [schoolViolence, sexualViolence, entryRestricted, loan, community, school_violence_help_1,
          school_violence_help_2, school_violence_education, sex_education, media_education,
          safety_education])

df = pd.merge(a, personal_edu, on=a.index)
df = df.rename(columns={'key_0':'시도'})
df = df.set_index('시도')

num_col = len(df.columns)
multi_col_1 = [f'T{j + 1}' for j in range(num_col)]
df.columns = multi_col_1
df
```
